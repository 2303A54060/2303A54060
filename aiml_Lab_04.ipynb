```python
l_column =list(df.columns)
len_feature = len(l_column)
l_column
```

```python
x=df[l_column[0:len_feature-2]]
y=df[l_column[len_feature-2]]
```

```python
print("Feature set size:",x.shape)
print("variable set size:",y.shape)


x.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train ,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=123)
```

```python
print("Traning feature set size:",x_train.shape)
print("Test feature set size:",x_train.shape)
print("Traning variable set size:",y_train.shape)
print("Test variable set size:",y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn import metrics
```

```python
lm = LinearRegression()
lm.fit(x_train,y_train)
```

```python
print("intercept term of the linear model:" , lm.intercept_)
```

```python
print("coefficients of the linear model:",lm.coef_)
```

```python
cdf = pd.DataFrame(data=lm.coef_, index=x_train.columns, columns=["Coefficients"])
cdf
```

```python
n = x_train.shape[0]
k = x_train.shape[1]
dfN = n - k

train_pred = lm.predict(x_train)
train_error = np.square(train_pred - y_train)
sum_error = np.sum(train_error)

se = [0,0,0,0,0]

for i in range(k):
  r = (sum_error / dfN)
  r = r / np.sum(np.square(x_train[list(x_train.columns)[i]] - x_train[list(x_train.columns)[i]].mean()))
  se[i] = np.sqrt(r)

cdf['Standard Error'] = se
cdf['t-Statistic'] = cdf['Coefficients'] / cdf['Standard Error']
cdf
```

```python
print("Therefore,features arranged  in the order of importance for predicting the house price\n",'-'*90,sep='')
l=list(cdf.sort_values('t-Statistic',ascending=False).index)
print(' > \n'.join(l))
```

```python
l=list(cdf.index)
from matplotlib import gridspec
fig = plt.figure(figsize=(18,10))
gs=gridspec.GridSpec(2,3)

ax0 = plt.subplot(gs[0])
ax0.scatter(df[l[0]], df['Price'])
ax0.set_xlabel(l[0] +"vs.price",fontdict={'fontsize':20})

ax1 = plt.subplot(gs[1])
ax1.scatter(df[l[1]],df['Price'])
ax1.set_xlabel(l[1] +"vs.price",fontdict={'fontsize':20})

ax2 = plt.subplot(gs[2])
ax2.scatter(df[l[2]],df['Price'])
ax2.set_xlabel(l[2] +"vs.price",fontdict={'fontsize':20})

ax3 = plt.subplot(gs[3])
ax0.scatter(df[l[3]],df['Price'])
ax0.set_xlabel(l[3] +"vs.price",fontdict={'fontsize':20})

ax0 = plt.subplot(gs[4])
ax0.scatter(df[l[4]],df['Price'])
ax0.set_xlabel(l[4] +"vs.price",fontdict={'fontsize':20})
```

```python
print("R-suared value of this fit:",round(metrics.r2_score(y_train,train_pred),3))
```

```python
predictions = lm.predict(x_test)

print("Type of the predicted object:",type(predictions))
print("size of the predictedd ov=bject:",predictions.shape)
```

```python
plt.figure(figsize=(10,7))
plt.title("Actual vs. predicted house  prices",fontsize=25)
plt.xlabel("Actual test set house prices",fontsize=18)
plt.ylabel("Predicted house prices",fontsize=18)
plt.scatter(x=y_test,y=predictions)
```
